```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_olivetti_faces
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
from matplotlib.ticker import MultipleLocator

def num_rand(rand_state, images, labels):
  global images_train, images_test, labels_train, labels_test
  images_train, images_test, labels_train, labels_test = train_test_split(
      images, labels, test_size=0.1, random_state=rand_state,
  )

def get_image_list(images_train):
    # 
    images_dict = {}

    # 
    for image, label in zip(images_train, labels_train):
        # 
        if label not in images_dict:
            images_dict[label] = []
        # 
        images_dict[label].append(image)

    # 
    images_list = []

    # 
    for key, value in images_dict.items():
        # 
        # value = [image1, image2, image3, image4, ...]
        # images_array = ndarray(num, height, width)
        images_array = np.array(value)
        # 
        images_list.append(images_array) #

    # [ndarray(num, height, width),
    # ndarray(num, height, width), ...]
    return images_list #

def csa(images_train, new_m, new_n, T_csa, kesai):
    # 
    m = images_train.shape[1]
    # 
    n = images_train.shape[2]
    old_U = np.identity(m)
    old_V = np.identity(n)

    for i_csa in range(T_csa):
        F = []
        G = []
        UUT = np.dot(old_U, old_U.T)
        VVT = np.dot(old_V, old_V.T)
        for i in range(images_train.shape[0]):
            X_i = images_train[i]
            X_i_U = np.dot(UUT, X_i)
            X_i_V = np.dot(X_i, VVT)
            F.append(X_i_U)
            G.append(X_i_V)
        F = np.hstack([X_i_U.T for X_i_U in F])
        FFT = np.dot(F, F.T)
        eigenvalues, eigenvectors = np.linalg.eigh(FFT)
        V = eigenvectors[:, 0:new_n]

        G = np.hstack([X_i_V for X_i_V in G])
        GGT = np.dot(G, G.T)
        eigenvalues, eigenvectors = np.linalg.eigh(GGT)
        U = eigenvectors[:, 0:new_m]

        if i_csa >= 1 and np.linalg.norm(U - old_U, ord='fro') < m * kesai and np.linalg.norm(V - old_V, ord='fro') < n * kesai:
            return U, V

        old_U = U
        old_V = V

    return U, V


def compute_scatter_V(images_train, U):
    # 
    X_mean = np.mean(images_train, axis=0)
    images_train = get_image_list(images_train) #
    S_b = 0
    S_w = 0
    # 
    for sample_class in images_train:
        # 
        n_c = len(sample_class)

        # 
        # [9,64,64] -> [9*64, 64]
        stacked_matrices = np.stack(sample_class)
        # 0
        X_mean_c = np.mean(stacked_matrices, axis=0)
        X_mean_cu = np.dot(U.T, X_mean_c)
        X_mean_u = np.dot(U.T, X_mean)
        temp = X_mean_cu - X_mean_u
        S_b += n_c * np.dot(temp.T, temp)

        for X_i in sample_class:
            X_iu = np.dot(U.T, X_i)
            S_w += np.dot((X_iu - X_mean_cu).T, (X_iu - X_mean_cu))

    return S_b, S_w


def compute_scatter_U(images_train, V):
    # 
    X_mean = np.mean(images_train, axis=0)

    images_train = get_image_list(images_train)

    S_b = 0
    S_w = 0

    # 
    for sample_class in images_train:
        # 
        n_c = len(sample_class)
        # 
        stacked_matrices = np.stack(sample_class)
        # 0
        X_mean_c = np.mean(stacked_matrices, axis=0)

        X_mean_cv = np.dot(X_mean_c, V)
        X_mean_v = np.dot(X_mean, V)
        temp = X_mean_cv - X_mean_v

        S_b += n_c * np.dot(temp, temp.T)

        for X_i in sample_class:
            X_iu = np.dot(X_i, V)
            S_w += np.dot((X_iu - X_mean_cv), (X_iu - X_mean_cv).T)

    return S_b, S_w


def dater(images_train, new_m, new_n, U_csa, V_csa, T_dater, kesai):
    # csa

    images_train = np.array([np.dot(U_csa.T, image).dot(V_csa) for image in images_train])

    # 
    m = images_train.shape[1]
    # 
    n = images_train.shape[2]

    U = np.identity(m)
    U = U[:, :new_m]

    old_U = U
    old_V = None

    # 
    for i in range(T_dater):
        # V
        S_b, S_w = compute_scatter_V(images_train, old_U)
        St = np.dot(np.linalg.inv(S_w), S_b)
        eigenvalues, eigenvectors = np.linalg.eigh(St)
        V = eigenvectors[:, -new_n:]

        # U
        S_b, S_w = compute_scatter_U(images_train, V)
        St = np.dot(np.linalg.inv(S_w), S_b)
        eigenvalues, eigenvectors = np.linalg.eigh(St)
        U = eigenvectors[:, -new_m:]

        if i >= 1 and np.linalg.norm(U - old_U, ord='fro') < m * kesai and np.linalg.norm(V - old_V, ord='fro') < n * kesai:
            return U, V

        old_U = U
        old_V = V

    return U, V

def custom_distance_metric(L_SP, L_SG):
    # Reshape the vectors back to matrices
    L_SP_matrix = L_SP.reshape(m_dater, n_dater)
    L_SG_matrix = L_SG.reshape(m_dater, n_dater)

    # Calculate the Frobenius norm
    distance = np.linalg.norm(L_SP_matrix - L_SG_matrix, ord='fro')

    return distance

def visualize_results1(accuracy_rank_1list, accuracy_rank_5list):
    T_values = [1, 2, 3, 4, 5]  # Iteration numbers
    fig, axs = plt.subplots(2, figsize=(4, 4))
    axs[0].plot(T_values, accuracy_rank_1list, marker='o')
    axs[0].set_title('Rank-1 Performance')
    axs[0].set_xlabel('Iteration Number')
    axs[0].set_ylabel('Rank-1 Recognition Rate')
    axs[0].legend()
    axs[0].grid(True)
    axs[0].xaxis.set_major_locator(MultipleLocator(1))
    axs[1].plot(T_values, accuracy_rank_5list, marker='o', color='green')
    axs[1].set_title('Rank-5 Performance')
    axs[1].set_xlabel('Iteration Number')
    axs[1].set_ylabel('Rank-5 Recognition Rate')
    axs[1].legend()
    axs[1].grid(True)
    axs[1].xaxis.set_major_locator(MultipleLocator(1))
    plt.tight_layout()
    plt.show()

def visualize_results2(dim_csa, accuracy_rank_1list, accuracy_rank_5list):
    dim_dater = [i for i in range(1,dim_csa)]
    fig, axs = plt.subplots(2, figsize=(4, 4))
    axs[0].plot(dim_dater, accuracy_rank_1list, marker=None)
    axs[0].set_title('Rank-1 Performance')
    axs[0].set_xlabel('Feature Dimension')
    axs[0].set_ylabel('Rank-1 Recognition Rate')
    axs[0].legend()
    axs[0].grid(True)
    axs[0].xaxis.set_major_locator(MultipleLocator(10))
    axs[1].plot(dim_dater, accuracy_rank_5list, marker=None, color='green')
    axs[1].set_title('Rank-5 Performance')
    axs[1].set_xlabel('Feature Dimension')
    axs[1].set_ylabel('Rank-5 Recognition Rate')
    axs[1].legend()
    axs[1].grid(True)
    axs[1].xaxis.set_major_locator(MultipleLocator(10))
    plt.tight_layout()
    plt.show()

dataset = fetch_olivetti_faces()
images = dataset.images
labels = dataset.target

for num in range(30,42):
  accuracy_rank_1list, accuracy_rank_5list = [], []
  print("set",num-29,":")
  for T in range(1,6):
    num_rand(num, images, labels)
    T_csa = T_dater = T
    # csamn
    m_csa = 60
    n_csa = 60
    # 
    kesai_csa = 1

    # datermn
    m_dater = 50
    n_dater = 50
    # 
    kesai_dater = 1
    U_csa, V_csa = csa(images_train,
                      new_m=m_csa,
                      new_n=n_csa,
                      T_csa=T_csa,
                      kesai=kesai_csa)
    U_dater, V_dater = dater(images_train,
                            new_m=m_dater,
                            new_n=n_dater,
                            U_csa=U_csa,
                            V_csa=V_csa,
                            T_dater=T_dater,
                            kesai=kesai_dater)

    U = np.dot(U_csa, U_dater)
    V = np.dot(V_csa, V_dater)
    images_train = np.array([np.dot(U.T, image).dot(V) for image in images_train])
    images_test = np.array([np.dot(U.T, image).dot(V) for image in images_test])
    images_train = images_train.reshape(images_train.shape[0], -1)
    images_test= images_test.reshape(images_test.shape[0], -1)
    # Create and train the KNN classifier
    knn = KNeighborsClassifier(metric=custom_distance_metric)
    # Use the original 3D shape for training and testing data
    knn.fit(images_train, labels_train)
    predictions = knn.predict(images_test)
    accuracy_rank_1 = accuracy_score(labels_test, predictions)
    neighbors = knn.kneighbors(images_test,n_neighbors=5, return_distance=False)
    accuracy_rank_5 = sum(1 for i, label in enumerate(labels_test) if label in labels_train[neighbors[i, :5]]) / len(labels_test)
    print(f"Rank-1 Accuracy: {accuracy_rank_1}")
    print(f"Rank-5 Accuracy: {accuracy_rank_5}")
    accuracy_rank_1list.append(accuracy_rank_1)
    accuracy_rank_5list.append(accuracy_rank_5)
  visualize_results1(accuracy_rank_1list, accuracy_rank_5list)

#plot Fig.4
print("Fig.4")
for num in [36,37]:
  for dim_csa in [60,55,50,45]:
    T_csa = T_dater = 3
    # csamn
    m_csa = dim_csa
    n_csa = dim_csa
    # 
    kesai_csa = 1
    # datermn
    accuracy_rank_1list, accuracy_rank_5list = [], []
    for dim_dater in range(1,dim_csa):
      num_rand(num, images, labels)
      m_dater = dim_dater
      n_dater = dim_dater
      # 
      kesai_dater = 1
      U_csa, V_csa = csa(images_train,
                        new_m=m_csa,
                        new_n=n_csa,
                        T_csa=T_csa,
                        kesai=kesai_csa)
      U_dater, V_dater = dater(images_train,
                              new_m=m_dater,
                              new_n=n_dater,
                              U_csa=U_csa,
                              V_csa=V_csa,
                              T_dater=T_dater,
                              kesai=kesai_dater)

      U = np.dot(U_csa, U_dater)
      V = np.dot(V_csa, V_dater)
      images_train = np.array([np.dot(U.T, image).dot(V) for image in images_train])
      images_test = np.array([np.dot(U.T, image).dot(V) for image in images_test])
      images_train = images_train.reshape(images_train.shape[0], -1)
      images_test= images_test.reshape(images_test.shape[0], -1)
      # Create and train the KNN classifier
      knn = KNeighborsClassifier(metric=custom_distance_metric)
      # Use the original 3D shape for training and testing data
      knn.fit(images_train, labels_train)
      predictions = knn.predict(images_test)
      accuracy_rank_1 = accuracy_score(labels_test, predictions)
      neighbors = knn.kneighbors(images_test,n_neighbors=5, return_distance=False)
      accuracy_rank_5 = sum(1 for i, label in enumerate(labels_test) if label in labels_train[neighbors[i, :5]]) / len(labels_test)
      accuracy_rank_1list.append(accuracy_rank_1)
      accuracy_rank_5list.append(accuracy_rank_5)
    print("set",num-29,",","dim_csa:","(",dim_csa,",",dim_csa,")")
    visualize_results2(dim_csa, accuracy_rank_1list, accuracy_rank_5list)
```

```python
# Train SVM classifier

from sklearn.svm import SVC

svm = SVC(kernel='linear', C=1.0, random_state=42)
svm.fit(images_train_pca, labels_train)

# Transform and predict on the test set
images_test_flat = images_test.reshape((len(images_test), -1))
images_test_scaled = scaler.transform(images_test_flat)
images_test_pca = pca.transform(images_test_scaled)
labels_pred_test = svm.predict(images_test_pca)

# Evaluate accuracy
accuracy = accuracy_score(labels_test, labels_pred_test) * 100
print(f"Face Recognition Accuracy: {accuracy:.2f}%")

# Display classification report and confusion matrix
print("\nClassification Report:")
print(classification_report(labels_test, labels_pred_test))

print("\nConfusion Matrix:")
print(confusion_matrix(labels_test, labels_pred_test))

# Display some predictions
_, axes = plt.subplots(1, 5, figsize=(12, 3))
for i, ax in enumerate(axes):
    ax.imshow(images_test[i], cmap='gray')
    ax.set_title(f"True: {labels_test[i]}\nPred: {labels_pred_test[i]}")
    ax.axis('off')

plt.show()

import os
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_olivetti_faces
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA

from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

# Evaluate accuracy
from sklearn.metrics import classification_report, confusion_matrix
import numpy as np

# Set the environment variable to avoid KMeans warning
os.environ['OMP_NUM_THREADS'] = '2'

# Load the Olivetti Faces dataset
dataset = fetch_olivetti_faces()
images = dataset.images
labels = dataset.target

# Define the number of sets
num_sets = 12

# Initialize lists to store accuracies
rank_1_accuracies = []
rank_5_accuracies = []

# Loop over sets
for i in range(num_sets):
    # Split the data into training and testing sets
    images_train, images_test, labels_train, labels_test = train_test_split(
        images, labels, test_size=0.5, random_state=i
    )

    # Flatten the images to use them as features in the SVM model
    n_samples, h, w = images_train.shape
    images_train_flat = images_train.reshape((n_samples, h * w))

    n_test, h, w = images_test.shape

    # Standardize the features
    scaler = StandardScaler()
    images_train_scaled = scaler.fit_transform(images_train_flat)

    # Reduce dimensionality using PCA
    n_components = 100
    pca = PCA(n_components=n_components, whiten=True).fit(images_train_scaled)
    images_train_pca = pca.transform(images_train_scaled)

    # Train SVM classifier
    from sklearn.svm import SVC

    svm = SVC(kernel='linear', C=1.0, random_state=42)
    svm.fit(images_train_pca, labels_train)

    # Transform and predict on the test set
    images_test_flat = images_test.reshape((len(images_test), -1))
    images_test_scaled = scaler.transform(images_test_flat)
    images_test_pca = pca.transform(images_test_scaled)
    labels_pred_test = svm.predict(images_test_pca)

    # Evaluate accuracy
    accuracy = accuracy_score(labels_test, labels_pred_test) * 100
    print(f"Set {i} - Face Recognition Accuracy: {accuracy:.2f}%")

    # Evaluate Rank-1 and Rank-5 accuracy
    rank_1_accuracy = np.sum(labels_test == labels_pred_test) / len(labels_test) * 100

    # Find the top-5 predictions for each test sample
    decision_function = svm.decision_function(images_test_pca)
    top_5_predictions = np.argsort(decision_function, axis=1)[:, ::-1][:, :5]

    # Check if the true label is in the top-5 predictions
    rank_5_accuracy = np.sum([label in top_5 for label, top_5 in zip(labels_test, top_5_predictions)]) / len(labels_test) * 100

    # Append accuracies to lists
    rank_1_accuracies.append(rank_1_accuracy)
    rank_5_accuracies.append(rank_5_accuracy)

    # Print accuracies for the current set
    print(f"Set {i} - Rank-1 Accuracy: {rank_1_accuracy:.2f}% | Rank-5 Accuracy: {rank_5_accuracy:.2f}%")
    print("-" * 50)

# Print overall Rank-1 and Rank-5 accuracy
print(f"\nOverall Rank-1 Accuracy: {np.mean(rank_1_accuracies):.2f}%")
print(f"Overall Rank-5 Accuracy: {np.mean(rank_5_accuracies):.2f}%")
```
